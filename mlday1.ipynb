## Pre-pruning 
We will see pre pruning techniques in action. The following are pre pruning Techniques.
1. Max depth
2. Max number of leaf nodes
3. Minimum impurity decrease


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1701)

x = np.random.randn(1000)/2
y = x**2+x**3 + 0.5*np.random.randn(1000)

plt.scatter(x, y)
plt.title("Random Dataset")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

###  Max depth
The tree grows until we reach the max depth limit. So we will fit a decision tree with max depth ranging from 1 to 6.

```python
from sklearn.tree import DecisionTreeRegressor
xp = np.arange(-2.0, 2.0, 0.01).reshape(-1, 1)
def plot_depth(d = 1):
    dt = DecisionTreeRegressor(max_depth = d)
    dt.fit(np.array(x).reshape(-1,1), y)
    yp =dt.predict(xp)
    plt.scatter(x, y, alpha=0.1)
    plt.plot(xp, yp, linewidth=2, color='red')
plt.figure(figsize = (20, 30))
for i in range(6):
    plt.subplot(3, 2, i+1)
    plot_depth(i+1)
    plt.title("Max Depth:"+str(i+1))
plt.show()
```

### Max no of leaf nodes
The tree grows until we reach the max leaf limit.So, We will fit a decision tree with different number of max leaves.

```python
def plot_leaf(d = 2):
    dt = DecisionTreeRegressor(max_leaf_nodes = d)
    dt.fit(np.array(x).reshape(-1,1), y)
    yp = dt.predict(xp)
    plt.scatter(x, y, alpha=0.1)
    plt.plot(xp, yp, linewidth=2, color='red')
plt.figure(figsize = (20, 30))
for i in range(6):
    plt.subplot(3, 2, i+1)
    plot_leaf(5*(i+1))
    plt.title("Max Leaf: "+str(5*(i+1)))
plt.show()
```

###  Min impurity decrease
The tree grows if a node induces a decrease of the impurity greater than or equal to this value.So, we will fit the model with different values.

```python
def plot_imp(d = 0.0) :
    dt = DecisionTreeRegressor(min_impurity_decrease = d)
    dt.fit(np.array(x).reshape(-1,1), y)
    yp = dt.predict(xp)
    plt.scatter(x, y, alpha=0.1)
    plt.plot(xp, yp, linewidth=2, color='red')    
plt.figure(figsize = (20, 30))
for i in range(6):
    plt.subplot(3, 2, i+1)
    plot_imp(((i+1)/100.0))
    plt.title("Min Impurity:"+str(((i+1)/100.0)))
plt.show()
```
